```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/Delinquency_prediction_dataset.csv');df
```

```python
df = df.drop('Customer_ID',axis=1);df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.hist(df['Income'])
print((df['Income'].isnull().sum() / df['Income'].count()) * 100 ) ##8.45 % missing data in Income
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True);
```

```python
plt.hist(df['Loan_Balance'])
print((df['Loan_Balance'].isnull().sum() / df['Loan_Balance'].count()) * 100 ) ##6.15 % missing data in Loan_Balance
```

```python
df['Loan_Balance'].fillna(df['Loan_Balance'].median(), inplace=True);
```

```python
df['Credit_Score'].fillna(df['Credit_Score'].median(), inplace=True);
```
